```python
# import packages
import requests
import pandas as pd
```

# Get tweets

```python
# read data
data = pd.read_csv("./application/dataset_service.tsv", sep='\t')
data.head(3)
```

```python
def get_data_json(raw_data):
    data_json = []
    for i in range(1, 100):
        
        data_json.append(raw_data.iloc[:, 1][i])

    return data_json
```

```python
get_data_json(data)
```

## Check prediction

```python
import json

url = 'http://127.0.0.1:8000/api/label'
data_json = {'data': get_data_json(data)}

response = requests.post(url, json=data_json, headers = {'content-type':'application/json'})
response.json()

    
```
